```python
"""
Question 2 - Recursive zigzag letter pattern.

The alphabet climbs from the BOTTOM row to the top, comes back down, and
repeats. Note from the question's examples that the top and bottom turns
each hold TWO consecutive letters (e.g. 'de' together on top, 'hi'
together on the bottom), so one full cycle uses 2 * rows letters - this
is NOT the classic rail-fence zigzag, whose period would be 2*(rows-1).
Letters wrap back to 'a' after 'z'. Each letter gets its own column.
"""

import sys


def build(grid, k, total, rows):
    """Recursively place letter k of the pattern into the grid.

    Args:
        grid:  2D character grid being filled (modified in place)
        k:     current position along the pattern (0 .. total-1)
        total: number of letters to place (2 * rows * cycles)
        rows:  number of rows in the grid
    """
    if k == total:                       # base case: all positions filled
        return
    m = k % (2 * rows)                   # position within the current cycle
    row = rows - 1 - m if m < rows else m - rows   # climb up, then come down
    grid[row][k] = chr(ord('a') + k % 26)          # wrap around after 'z'
    build(grid, k + 1, total, rows)      # recursive call


def print_rows(grid, r):
    """Recursively print the grid, one row per call (top row first)."""
    if r == len(grid):                   # base case: no rows left
        return
    print(''.join(grid[r]).rstrip())     # rstrip only trims trailing spaces
    print_rows(grid, r + 1)              # recursive call


def zigzag(rows, cycles):
    """Validate the inputs, then build and print the zigzag pattern."""
    if rows < 1:
        print('Error: rows must be at least 1.')
        return
    if cycles < 1:
        print('Error: cycles must be at least 1.')
        return

    total = 2 * rows * cycles            # each cycle uses 2 * rows letters
    # build() recurses once per letter; raise the limit for large inputs.
    # (For very large patterns, an iterative loop would avoid deep recursion.)
    sys.setrecursionlimit(max(total + 100, 1000))

    grid = [[' '] * total for _ in range(rows)]
    build(grid, 0, total, rows)
    print_rows(grid, 0)


rows = int(input("rows: "))
cycles = int(input("cycles: "))
zigzag(rows, cycles)
```
