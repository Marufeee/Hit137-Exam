```python
import sys


def build(grid, k, total, rows):
    if k == total:
        return
    period = 2 * (rows - 1)
    m = k % period
    row = m if m < rows else period - m
    grid[row][k] = chr(ord('a') + k % 26)
    build(grid, k + 1, total, rows)


def print_rows(grid, r):
    if r == len(grid):
        return
    print(''.join(grid[r]).rstrip())
    print_rows(grid, r + 1)


def zigzag(rows, cycles):
    if rows < 2:
        print('Error: rows must be at least 2.')
        return
    if cycles < 1:
        print('Error: cycles must be at least 1.')
        return
    period = 2 * (rows - 1)
    total  = period * cycles
    sys.setrecursionlimit(total + rows + 100)
    grid = [[' '] * total for _ in range(rows)]
    build(grid, 0, total, rows)
    print_rows(grid, 0)


rows   = int(input('rows: '))
cycles = int(input('cycles: '))
zigzag(rows, cycles)
```
